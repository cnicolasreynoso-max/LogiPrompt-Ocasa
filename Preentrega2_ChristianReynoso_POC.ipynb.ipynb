```python
import json

def logiprompt_engine_mock(mensaje_chofer):
    """
    POC LogiPrompt: Simulación de respuesta de IA (Fast Prompting).
    Esta función emula el comportamiento de gpt-3.5-turbo para demostrar
    la lógica de la solución sin requerir saldo en la API.
    """
    mensaje = mensaje_chofer.lower()
    
    # Simulación de la lógica de Few-Shot Prompting
    if "no pude" in mensaje or "no me atendieron" in mensaje or "no atendieron" in mensaje:
        resultado = {
            "incidente": "AVISO_NO_CONSEGUIDO",
            "protocolo_remito_cabina": True,
            "prioridad": "BAJA",
            "accion_sugerida": "Dejar aviso de visita físico y continuar ruta."
        }
    elif "remito" in mensaje or "papeles" in mensaje or "documento" in mensaje:
        resultado = {
            "incidente": "SIN_REMITO",
            "protocolo_remito_cabina": False,
            "prioridad": "ALTA",
            "accion_sugerida": "Detener descarga. Buscar documento o solicitar reemisión."
        }
    else:
        resultado = {
            "incidente": "TURNO_FALLIDO",
            "protocolo_remito_cabina": True,
            "prioridad": "MEDIA",
            "accion_sugerida": "Notificar a oficina para reprogramar turno con Molinos."
        }
    
    return json.dumps(resultado, indent=4, ensure_ascii=False)

# --- DEMOSTRACIÓN INTERACTIVA ---
print("--- SISTEMA LOGIPROMPT OCASA: ASISTENTE DE ÚLTIMA MILLA ---")
entrada = input("Ingrese el reporte del chofer: ")

print("\n[PROCESANDO CON IA (SIMULACIÓN FAST PROMPTING)...]")
print(logiprompt_engine_mock(entrada))
```
